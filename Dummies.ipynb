En el siguiente código se encarga de procesar los géneros de los juegos del DataFrame llamado 'df_games'.  Donde :   
* Expande la columna 'genres' para tener una fila por cada género, luego crea variables ficticias (dummies) para cada género y suma por el ID del juego.

* Combina el ID del juego con las variables ficticias de género. Posteriormente, identifica las columnas que se mantendrán, excluyendo aquellas que contienen "&amp;" en cualquier caso. 

* Filtra el DataFrame para mantener solo las columnas deseadas. Por último, se asegura de que los nombres de las columnas sean únicos y devuelve el DataFrame resultante con los géneros y los ID de los juegos.


```python
import pandas as pd


# Assuming your DataFrame is named 'df_games'

# Explode genres, create dummies, and sum by game id
genres_dummies = (
    df_games['genres'].explode()
    .str.get_dummies()
    .groupby(level=0)
    .sum()
    .reset_index(drop=True)
)

# Combine game id with genre dummies
all_genres = pd.concat([df_games[['id']], genres_dummies], axis=1)

# Identify columns to keep (excluding those with any case of "&amp;")
keep_cols = all_genres.columns[~all_genres.columns.str.contains('&amp;', case=False)]

# Filter DataFrame to keep desired columns
all_genres = all_genres[keep_cols]

# Ensure unique column names (if necessary)
all_genres.columns = pd.unique(all_genres.columns)

all_genres
```

```python
all_genres.to_parquet(r'C:\Users\Usuario\Desktop\proyecto steam\df_dummies.parquet', index=False)
```
